```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
import string
from nltk.tokenize import word_tokenize

```

```python
data=pd.read_csv('/Users/rishismac/Desktop/Twitter_sentiment_train.csv')
data.isna().sum()
```

```python
xtrain=list(data.text)
xtrain
ytrain=list(data.airline_sentiment)
ytrain
```

```python
from nltk.corpus import stopwords
stop=stopwords.words('english')
stop
```

```python
punctuations=list(string.punctuation)
stop=stop+punctuations
stop
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize('better','a')
```

```python
pos_tag(['better'])
```

```python
def simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
```

```python
documents=[]
for text in xtrain:
    words=[lemmatizer.lemmatize(w,simple_pos(pos_tag([w])[0][1])) for w in word_tokenize(text) if w not in stop]
    documents.append(words)

```

```python
documents
```

```python
a=['aa','bb']
' '.join(a)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer(max_df=.99,min_df=0.001)
text_documents=[' '.join(file) for file in documents]
text_documents
```

```python
from sklearn.model_selection import train_test_split
split_xtrain,split_xtest,split_ytrain,split_ytest=train_test_split(text_documents,ytrain)
```

```python
xtrain_df=count_vec.fit_transform(split_xtrain)
xtrain_df.todense()
xtrain_df.shape
```

```python
count_vec.get_feature_names()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

svc=SVC()
grid={'C':[1e1,1e2,1e3,1e4,1e5],
     'gamma':[1e1,1,1e-1,1e-2,1e-3]}
abc=GridSearchCV(svc,grid)
abc.fit(xtrain_df,split_ytrain)
```

```python
abc.best_estimator_

#grid search cv gives the best hyper parameters as C=10.0, gamma=0.01
```

```python
from sklearn.svm import SVC
svc=SVC(C=10.0, gamma=0.01)
svc.fit(xtrain_df,split_ytrain)
```

```python
xtest_df=count_vec.transform(split_xtest)
```

```python
svc.score(xtest_df,split_ytest)
```

## Trying MultiNomial Naive Bayes

```python

from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(xtrain_df,split_ytrain)
clf.score(xtest_df,split_ytest)
#score obtained is less than SVC 
```

## Trying the KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
kclf=KNeighborsClassifier()
kclf.fit(xtrain_df,split_ytrain)
kclf.score(xtest_df,split_ytest)
```

## Trying the Random Forest Classifer

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(xtrain_df,split_ytrain)
rfc.score(xtest_df,split_ytest)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain_df,split_ytrain)
lr.score(xtest_df,split_ytest)
```
